```python
import pandas as pd
#Reading Files
ratings_df = pd.read_csv('ratings.csv')
movies_df = pd.read_csv('movies.csv')
```

```python
movrat_df=pd.merge(ratings_df,movies_df,how='inner',on='movieId')
```

```python
mean_rat=movrat_df[['title','rating']].groupby(['title']).mean().sort_values(by=['rating'],ascending=False)
```

```python
count_rat=movrat_df[['title','rating']].groupby(['title']).count().sort_values(by=['rating'],ascending=False)
```

```python
mean_rat=mean_rat.rename(columns={'rating':'mean_rating'})
count_rat=count_rat.rename(columns={'rating':'count_rating'})
```

```python
countmean_rat=pd.merge(mean_rat,count_rat,how='inner',on='title')
countmean_rat.sort_values(by=['count_rating','mean_rating'], ascending=False)
```

```python
df_merge=pd.merge(ratings_df,movies_df,how='inner',on='movieId')
new_df=df_merge.pivot_table(index='userId',columns='title',values='rating')
```

```python
new_df
```

```python
serie_bm=pd.Series(data=new_df['Forrest Gump (1994)'].values)
```

```python
result=new_df.corrwith(serie_bm)
```

```python
corr_df=pd.DataFrame(data=result.values,columns=['Correlaciones'], index=result.index).dropna()
print(corr_df)
```

```python
corr_count=corr_df.join(count_rat)
```

```python
recomend=corr_count.sort_values(by=['Correlaciones','count_rating'], ascending=False)
recomend=recomend.astype('float')
```
